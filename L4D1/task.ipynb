```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("reduced.jpg", 0)


def reduce(img, levels):
    step = 256 // levels
    return (img // step) * step

#levels
img4 = reduce(img, 4)
img8 = reduce(img, 8)
img16 = reduce(img, 16)

plt.figure(figsize = (10,4))

# original
plt.subplot(1,4,1)
plt.imshow(img, cmap = 'gray')
plt.title("Original")
plt.axis("off")

# 4
plt.subplot(1,4,2)
plt.imshow(img4, cmap = 'gray')
plt.title("4 levels")
plt.axis("off")

# 8
plt.subplot(1,4,3)
plt.imshow(img8, cmap = 'gray')
plt.title("8 levels")
plt.axis("off")

# 
plt.subplot(1,4,4)
plt.imshow(img16, cmap = 'gray')
plt.title("16 levels")
plt.axis("off")
          

plt.show()
```
