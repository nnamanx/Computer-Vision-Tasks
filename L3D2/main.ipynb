```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img1_path = "bird.jpeg"
img2_path = "sun2.jpeg"

img1 = cv2.imread(img1_path, cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread(img2_path, cv2.IMREAD_GRAYSCALE)

if img1.shape != img2.shape:
    print("Resizing happens the first...")
    img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

added_image = cv2.add(img1, img2)  

average_image = ((img1.astype(np.float32) + img2.astype(np.float32)) / 2).astype(np.uint8)

plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.imshow(img1, cmap='gray')
plt.title("Image 1")
plt.axis('off')

plt.subplot(2,2,2)
plt.imshow(img2, cmap='gray')
plt.title("Image 2")
plt.axis('off')

plt.subplot(2,2,3)
plt.imshow(added_image, cmap='gray')
plt.title("Added Image")
plt.axis('off')


plt.subplot(2,2,4)
plt.imshow(average_image, cmap='gray')
plt.title("Average Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```
