```python
import numpy as np
from imageio import imread
import matplotlib.pyplot as plt

# (H, W, 3)
img = imread('reduced.jpg') / 255.0

weights = np.array([0.299, 0.587, 0.114])
gray_weighted = np.dot(img, weights)

gray_average = img.mean(axis=2)

difference = np.abs(gray_weighted - gray_average)

plt.figure(figsize=(16, 6))

# slide formula 
plt.subplot(2, 3, 1)
plt.imshow(gray_weighted, cmap = 'gray')
plt.title('Weighted Grayscale')
plt.axis('off')

# Average 
plt.subplot(2, 3, 2)
plt.imshow(gray_average, cmap ='gray')
plt.title('Average Grayscale')
plt.axis('off')

# Difference 
plt.subplot(2, 3, 3)
plt.imshow(difference, cmap = 'hot')
plt.title('Difference (|Weighted - Average|)')
plt.axis('off')

# Histogram 
plt.subplot(2, 1, 2)
plt.hist(gray_weighted.ravel(), bins = 256, color = 'blue', alpha = 0.5, label = 'Weighted')
plt.hist(gray_average.ravel(), bins = 256, color = 'red', alpha = 0.5, label = 'Average')
plt.title('Pixel Intensity Histogram')
plt.xlabel('Intensity (0-1)')
plt.ylabel('Number of pixels')
plt.legend()

plt.tight_layout()
plt.show()
```
